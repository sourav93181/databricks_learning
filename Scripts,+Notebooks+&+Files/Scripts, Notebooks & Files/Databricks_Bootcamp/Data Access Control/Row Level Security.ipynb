### **Mapping Table**

```python
%sql
CREATE TABLE databricksansh.silver.rls_mapping
(
  user_id STRING,
  region STRING 
)
```

```python
%sql
INSERT INTO databricksansh.silver.rls_mapping 
VALUES
("anshlambaaz@gmail.com","east"),
("lovelamba@gmail.com","west"),
("lovelovelamba@gmail.com",'south'),
('lovemelamba@gmail.com','north')
```

```python
%sql
SELECT EXISTS
(SELECT * FROM databricksansh.silver.rls_mapping
WHERE 
user_id = current_user() AND
region = 'west')
```

### **CREATE RLS FUNCTION**

```python
%sql
CREATE OR REPLACE FUNCTION databricksansh.silver.rls_func(p_region STRING)
RETURNS BOOLEAN 
RETURN 
EXISTS
(SELECT * FROM databricksansh.silver.rls_mapping
WHERE 
user_id = current_user() AND
region = lower(p_region))
```

### **APPLYING FUNCTION ON REGION COLUMN**

```python
%sql
ALTER TABLE databricksansh.silver.stores_enr
SET ROW FILTER databricksansh.silver.rls_func ON (region)
```
