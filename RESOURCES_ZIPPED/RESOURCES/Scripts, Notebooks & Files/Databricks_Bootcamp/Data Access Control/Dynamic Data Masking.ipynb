### **CREATE THE MASK FUNCTIONS**

```python
%sql
CREATE OR REPLACE FUNCTION databricksansh.silver.dynamic_mask(p_email STRING)
RETURN CASE WHEN is_account_group_member('developers') THEN p_email ELSE '***' END;
```

### **APPLYING MASK**

```python
%sql
ALTER TABLE databricksansh.silver.customers_enr
ALTER COLUMN email SET MASK databricksansh.silver.dynamic_mask
```
